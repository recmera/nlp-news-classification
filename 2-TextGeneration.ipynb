### Generacion de noticias

```python
import pandas as pd
import numpy as np
#import transformers
#from transformers import pipeline
#transformers.__version__
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("dataEtiquetada.csv")
df.drop(columns=['media_outlet','url','title','content', 'topic'], inplace=True)
print(len(df))
display(df['label'].value_counts())
```

```python
df3 = df.loc[df['label']=="ciencia y tecnología"] #tope

df1 = df.loc[df['label']=="economía"]
df4 = df.loc[df['label']=="catástrofes y accidentes"]
df5 = df.loc[df['label']=="cultura y arte"]
df7 = df.loc[df['label']=="ecología y planeta"]
df8 = df.loc[df['label']=="crimen, delitos y justicia"]
df9 = df.loc[df['label']=="salud"]
```

```python
display(df7.head(4), df5.head(4),df4.head(4) )
```

#### 1. Generación de noticias categoría *7.ecología y planeta*

```python
import random 
togen = len(df3) - len(df7)
text = df7[['text']]
#text_generator = pipeline("text-generation", model="DeepESP/gpt2-spanish", tokenizer="DeepESP/gpt2-spanish")
generated_cat7 = []

for i in range(togen):
    string1 = str(text.values[random.randint(0, len(df7)-1)]).strip("['']")[0:random.randint(0,len(text))]
    string2 = str(text.values[random.randint(0, len(df7)-1)]).strip("['']")[0:random.randint(0,len(text))]
    string3 = str(text.values[random.randint(0, len(df7)-1)]).strip("['']")[0:random.randint(0,len(text))]
    res = string1 + string2 + string3
    generated_cat7.append(res)
    #generated_cat7.append(text_generator(res, max_length=100, do_sample=False))
```

```python
cat7 = pd.DataFrame(generated_cat7,  columns = ['text'])
cat7['label']= "ecología y planeta"
```

```python
cat7
```
